```python
import plotly.graph_objects as go
import plotly.offline as py
import plotly.io as pio
import matplotlib.pyplot as plt
import numpy as np

from plotly.subplots import make_subplots
from scipy.stats import norm

pio.templates.default = "simple_white"
```

```python
x = np.arange(-4,4,0.001)
gaussian_ndist = norm.pdf(x)
random_ndist = norm.rvs(scale=1,loc=5, size=1000)
pbDensity_random =  norm.pdf(random_ndist) #norm(loc=5, scale=1).pdf(x)


fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        'Analytic Gaussian (normal) distribution',
        'Analytic Gaussian (normal) distribution',
        'Empirical normal distribution',''))

fig.add_trace(go.Scatter(x=x, y=gaussian_ndist,
                    mode='lines',
                    line=dict(color="dodgerblue",width=2)
                    ),row=1, col=1)

fig.add_trace(go.Scatter(x=x, y=gaussian_ndist,
                    mode='lines',
                    line=dict(color="dodgerblue",width=2)
                    ),row=1, col=2)

fig.add_trace(go.Histogram(x=random_ndist,
                    marker_color = 'dodgerblue'
                    ),row=2, col=1)

fig.add_trace(go.Histogram(x=pbDensity_random,
                    marker_color = 'dodgerblue', nbinsx=25
                    ),row=2, col=2)


# for row in [1,2]:
#     for column in [1,2]:

#         fig.update_xaxes(
#             dtick=1, row=row, col=column)


fig.update_layout(
    showlegend=False, 
    height=700, width=1200, 
    title_text="Unsupervising learning challanges. Chapter:2")

fig.show()
```
