```python
import math
import numpy as np
import plotly.graph_objects as go
import plotly.offline as py
from plotly.subplots import make_subplots
import plotly.io as pio

pio.templates.default = "plotly_dark"
```

```python
x = np.arange(0, 10, 0.1)

fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=("Linear y-axis", "Logarithmic y-axis"))

fig.add_trace(go.Scatter(x=x, y=x,
                    mode='lines',
                    line=dict(color="#4361ee",width=5),
                    name='x = x'),row=1, col=1)

fig.add_trace(go.Scatter(y=[math.e**i for i in x], x=x,
                    mode='lines',
                    line=dict(color="#3CAAE3",width=5),
                    name='x=e**x'),row=1, col=1)

fig.add_trace(go.Scatter(y=x, x=x,
                    mode='lines',
                    line=dict(color="#4361ee",width=5),
                    name='x = x'),row=1, col=2)

fig.add_trace(go.Scatter(y=[math.e**i for i in x], x=x,
                    mode='lines',
                    line=dict(color="#3CAAE3",width=5),
                    name='x=e**x'),row=1, col=2)

fig.update_yaxes(
    range=[0,100], 
    showgrid=False,
    showline=True, linewidth=1, linecolor='white',
    row=1, col=1)

fig.update_xaxes(
    showticklabels=False,
    showgrid=False,
    dtick=False,
    showline=True, linewidth=1, linecolor='white',
    row=1, col=1)

fig.update_yaxes(
    type='log',
    range=[-2,2], 
    showgrid=False,
    showline=True, linewidth=1, linecolor='white',
    row=1, col=2)

fig.update_xaxes(
    showticklabels=False,
    showgrid=False,
    dtick=False,
    showline=True, linewidth=1, linecolor='white',
    row=1, col=2)

fig.update_layout(showlegend=False)


fig.update_layout(height=600, width=1200, title_text="Unsupervising learning challanges. Chapter:1")
fig.show()
```
